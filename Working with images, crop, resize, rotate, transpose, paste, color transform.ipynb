## pasting one image to another image

```python
from PIL import Image

img1 = Image.open('Bishal Gautam.jpg')
img2 = Image.open('img1_small.jpg')
# img2_resized = img2.resize(400, 400)
dimension = (0, 0, 400, 400)
img2_cropped = img2.crop(dimension)
img_pasted = img1.paste(img2_cropped, dimension)
img_pasted.save('img_pasted.jpg', 'JPEG')
```

#### To paste images to file and remove error that was caused because of text file we now check the extension

```python
import os
from PIL import Image

img1 = Image.open('img1_small.jpg') # The dimension must match with image dimension and can be checked using img.size
dimension = (500,500,600,600) # give the location of image to be pasted
for i, file in enumerate(os.listdir(), 1):
    ext = os.path.splitext(file)
    if ext[-1].lower() in ['jpg', 'png', 'gif']:
        img2 = Image.open(file)
        img2.paste(img1, dimension)
        img2.save(f'pasted_image_{i}{ext[-1]}', 'JPEG')
```
