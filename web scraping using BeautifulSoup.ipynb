```python
from bs4 import BeautifulSoup
import requests

import csv

source = requests.get('https://wsvincent.com/').text

soup = BeautifulSoup(source, 'lxml')

csv_file = open('new_csvfile.csv', 'w')
csv_file_writer = csv.writer(csv_file)
csv_file_writer.writerow(['SN', 'Books','Remarks'])

i = 1

for content in soup.find_all('a', class_='post-link'):

    print()
    print(content.text)
    print()
    csv_file_writer.writerow([i, content.text, 'Best ever remarks'])
    i += 1

csv_file.close()

with open('new_csvfile.csv', 'r') as csv_file:
    csv_file_reader = csv.reader(csv_file)
    print(csv_file_reader)
    
    for content in csv_file_reader:
        print(content)
```
